# GR Framework in Julia

## 2D Plotting

```julia
using GR
```

```julia
X = LinRange(-2, 2, 40)
Y = X.^3 + X.^2 + X

plot(X,Y)
```

## Heatmap Plotting

```julia
x = randn(100000)
y = randn(100000)

hexbin(x, y)
```

## 3D Plotting

```julia
x = LinRange(0, 30, 1000)
y = cos.(x) .* x
z = sin.(x) .* x

plot3(x, y, z)
```
