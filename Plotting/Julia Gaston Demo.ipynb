# Julia Gnuplot Wrapper: Gaston

```julia
using Gaston
using LaTeXStrings

X = range(-2pi, stop=2pi, length=100)
Y = sin.(0.3 .+ 0.7X)

plotcom = ""
figurename = "JuliaGnuplot"

plot(X, Y, font="Serif,16", plotstyle="lines", title=raw"Title $f=f(x)$" ,legend=raw"Function $f=f(x)$", gpcom=plotcom)
```
